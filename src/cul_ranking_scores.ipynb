```python
import trueskill
import json
```

```python
with open('rankings.json', 'r') as file:
    matches = json.load(file)
```

```python
len(matches)
```

```python
env = trueskill.TrueSkill(mu=1000, sigma=333, draw_probability=0)
env.make_as_global()

ratings = {}
```

```python
for match in matches:
    if len(match) == 1:
      print(match)
      continue
    current_ratings = [[ratings.get(player, trueskill.Rating())] for player in match]
    ranks = list(range(len(match)))

    new_ratings = env.rate(current_ratings, ranks=ranks)

    for player, rating in zip(match, new_ratings):
        ratings[player] = rating[0]
```

```python
ratings
```

```python
sorted_ratings = sorted(ratings.items(), key=lambda x: x[1].mu, reverse=True)
sorted_ratings
```

```python
sorted_ratings_json = [{'player': player, 'mu': rating.mu, 'sigma': rating.sigma} for player, rating in sorted_ratings]
```

```python
with open('player_ratings.json', 'w') as json_file:
    json.dump(sorted_ratings_json, json_file, indent=4)
```
